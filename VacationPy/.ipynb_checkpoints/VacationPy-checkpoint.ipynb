```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint
from IPython.display import HTML
from ipywidgets.embed import embed_minimal_html

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
cities = pd.read_csv("output_data/cities.csv", encoding="utf-8")
cities.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
humidity = cities["Humidity"]
max_humidity = humidity.max()
locations = cities[["Lat", "Lng"]]
```

```python
fig = gmaps.figure(center=(20.0, 0.0), zoom_level=2)
heat_layer = gmaps.heatmap_layer(locations, weights=humidity,dissipating=False, 
                                 max_intensity=max_humidity,point_radius=1)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
ideal_city_df = cities.loc[(cities["Max Temp"] > 65) & (cities["Max Temp"] < 70) & (cities["Cloudiness"] == 0), :]
ideal_city_df = ideal_city_df.dropna(how='any')
ideal_city_df.reset_index(inplace=True)
del ideal_city_df['index']
ideal_city_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_list =[]

for i in range(len(ideal_city_df)):
    lat = ideal_city_df.loc[i]['Lat']
    lng = ideal_city_df.loc[i]['Lng']

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    params = {
        "location": f"{lat},{lng}",
        "radius": 5000,
        "type": "hotel",
        "key": g_key,
    }

    response = requests.get(base_url, params=params).json()
#     pprint(response)
    try:
        hotel_list.append(response['results'][0]['name'])
    except:
        hotel_list.append("")

ideal_city_df["Hotel Name"] = hotel_list
hotel_df = ideal_city_df.dropna(how='any')
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations)
fig.add_layer(markers)

# Display Map
fig
```
